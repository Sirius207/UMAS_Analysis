```python
# Extract Datetime (2014-01-01 00:00:00) to datetime features (year,mon, day, hr, min, weekday, is_holiday))

def add_time_features(house):
    house['datetime'] = house['Date & Time'].map(lambda x: datetime.strptime(x[:19], "%Y-%m-%d %H:%M:%S"))
    house = house.sort_values(by=['datetime'])
    house['year'] = [time.year for time in house['datetime']]
    house['mon'] = [time.month for time in house['datetime']]
    house['day'] = [time.day for time in house['datetime']]
    house['hr'] = [time.hour for time in house['datetime']]
    house['min'] = [time.minute for time in house['datetime']]
    house['weekday'] = [time.weekday() for time in house['datetime']]
    
    is_holiday_format = lambda x: 1 if x>4 else 0
    house['is_holiday'] = house['weekday'].map(is_holiday_format)
    
    # 3600 * 13 -> GMT-5 ~ GMT+8
    house['time'] = [int(datetime.timestamp(time)) + 3600*13 for time in house['datetime']]
    
    house= house.drop(['datetime'], axis=1)

    return house

power_df = pd.read_csv(f'./data/Home{house_id}/{year}/Home{house_id}-meter{meter_number}_{year}.csv')
time_power_df = add_time_features(power_df)
time_power_df.head()
```

```python
def show_col_counts(time_power_df, col):
    if col in time_power_df.columns:
        print(f"valid {col} column: ", len(time_power_df[time_power_df[col] != 0]))
    

def show_valid_col_counts(time_power_df, col):
    post_list = ['', '_x', '_y']
    
    for post_fix in post_list:
        show_col_counts(time_power_df, f"{col}{post_fix}")
    

def explore_merge_df(time_power_df):
    use = 'use [kW]'
    gen = 'gen [kW]'
    solar = 'Solar [kW]'
    overall = 'House overall [kW]'
    
    show_valid_col_counts(time_power_df, use)
    show_valid_col_counts(time_power_df, gen)
    show_valid_col_counts(time_power_df, solar)
    show_valid_col_counts(time_power_df, overall)
```

```python
def merge_process_single_meter(house_id, year, meter_id):
    weather_df = pd.read_csv(f'./data/Home{house_id}/home{house_id}{year}.csv')
    power_df = pd.read_csv(f'./data/Home{house_id}/{year}/Home{house_id}-meter{meter_id}_{year}.csv')
    
    time_power_df = add_time_features(power_df)
    merge_df = pd.merge(time_power_df,weather_df, on=['time'],how='outer')
    merge_df = fill_nan_weather_data(merge_df)

    # save files
    time_power_df.to_csv(f'./data/Home{house_id}/preprocess/home{house_id}_{year}_time_{meter_id}.csv', index=None)
    merge_df.to_csv(f'./data/Home{house_id}/preprocess/home{house_id}_{year}_merge_{meter_id}.csv', index=None)
    
    return time_power_df, merge_df


house_id = 'F'
meter_id = 2
time_power_df, merge_df = merge_process_single_meter(house_id, 2014, meter_id)
merge_process_single_meter(house_id, 2015, meter_id)

merge_df[['Date & Time', 'time','temperature','humidity']][44:64]
```

```python
# Main Function -> merge an year whole meter

def merge_process(house_id, year):
    weather_df = pd.read_csv(f'./data/Home{house_id}/home{house_id}{year}.csv')
    
    meters = [1,2,3,4]
    
    start = 0
    for meter_id in meters:
        try:
            power_df = pd.read_csv(f'./data/Home{house_id}/{year}/Home{house_id}-meter{meter_id}_{year}.csv')
            
            start += 1
            if start == 1:
                 all_power_df = power_df
            else:
                 all_power_df = pd.merge(all_power_df, power_df, on=['Date & Time'],how='inner')
    
            
        except FileNotFoundError:
            print(f'{house_id} - {year} - no meter {meter_id}')
    
    time_power_df = add_time_features(all_power_df)
    
    # for house C only
    time_power_df['total gen'] = time_power_df['gen [kW]_x'] + time_power_df['gen [kW]_y']
    
    # merge power data & weather data
    merge_df = pd.merge(time_power_df,weather_df, on=['time'],how='inner')
    
    # save files
    time_power_df.to_csv(f'./data/Home{house_id}/preprocess/home{house_id}_{year}_time_all.csv', index=None)
    merge_df.to_csv(f'./data/Home{house_id}/preprocess/home{house_id}_{year}_merge_all.csv', index=None)
    
    return time_power_df, merge_df


# 30min -> 17520 data a year
# 1hr -> 8760 data a year
# 1min -> 525600 data a year

house_id = 'C'
year = '2014'
time_power_df, merge_df = merge_process(house_id, year)
print('ok')

print("Time Shape", time_power_df.shape)
print("Merge Shape", merge_df.shape)
print("Merge Columns", merge_df.columns)

explore_merge_df(merge_df)

time_power_df.head()
```

```python
# Main function Concate Each Year Data


years = [2014, 2015, 2016]
house_id = 'C'


def concat_df(house_id, years):
    time_power_df_list = []
    merge_df_list = []
    
    for year in years:
        try:
            time_power_df, merge_df = merge_process(house_id, year)

            time_power_df_list.append(time_power_df)
            merge_df_list.append(merge_df)
        except FileNotFoundError:
            print(f'{house_id} - no year {year}')
    
    # concat whole years
    group_time_power_df = pd.concat(time_power_df_list, axis=0)
    group_merge_df = pd.concat(merge_df_list, axis=0)
    
    # save data
    group_time_power_df.to_csv(f'./data/Home{house_id}/preprocess/home{house_id}_time.csv', index=None)
    group_merge_df.to_csv(f'./data/Home{house_id}/preprocess/home{house_id}_merge.csv', index=None)

print('start concat...')
concat_df(house_id, years)
print('concat ok')
```
